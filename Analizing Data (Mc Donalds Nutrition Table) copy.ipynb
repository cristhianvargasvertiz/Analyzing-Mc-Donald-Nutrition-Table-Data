```python
import pandas as pd
import sqlite3

# Leer datos desde el archivo CSV
# El csv se descargo de https://www.kaggle.com/datasets/mcdonalds/nutrition-facts?resource=download
# El archivo se guardo en el directorio del script Proyecto SQL
data = pd.read_csv('./menu.csv')

# Conectar con la base de datos SQLite
conn = sqlite3.connect('McDonalds.db')

# Guardar los datos en la tabla 'MC_DONALDS_NUTRITION' en la base de datos SQLite
data.to_sql('MC_DONALDS_NUTRITION', conn, index=False, if_exists='replace')

# Leer los datos de la tabla 'MC_DONALDS_NUTRITION' en un DataFrame
df = pd.read_sql('SELECT * FROM MC_DONALDS_NUTRITION', conn)

# Imprimir el DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns

# Categorical Scatter Plots (Grafico de Puntos) VALORES COLUMNA SODIUM vs CATEGORIA COLUMNA x (Cada Fila o observacion) 

plot = sns.swarmplot(x='Category', y='Sodium', data=df) 
plt.setp(plot.get_xticklabels(), rotation=70)
plt.title('Sodium Content')
plt.show()
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns

# Categorical Scatter Plots (Grafico de Puntos) VALORES COLUMNA SODIUM vs CATEGORIA COLUMNA x (Cada Fila o observacion)

# Crea el grafico Scatter Plots estableciendo los ejes, (hue=) para que los puntos sea de colores
plot = sns.swarmplot(x='Category', y='Sodium', hue='Category', data=df) 

# plt.setp hace setup del grafico
plt.setp(plot.get_xticklabels(), rotation=70)

# Pone el titulo
plt.title('Sodium Content')
plt.show()
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns

# Values Scatter Plots (Grafico de Puntos) VALORES COLUMA1 vs VALORES COLUMNA2 X CADA FILA (o cada observacion)
# sns.jointplot agrega histogramas marginales a cada margen del grafico en los sentido del eje
# Los Histogramas son graficos que marcan valores en intervalos ordenados de mayor a menor ejemplo: (0,100), (100,200), (200,300) ... 

plot = sns.jointplot(x='Protein', y='Total Fat', data=df)

# no es necesario usar plot.show() debido a que esta ejecutando la linea magica de matplotlib inline
```
